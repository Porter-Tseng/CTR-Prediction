## <span style="color:skyblue; font-weight:bold;">Data Preprocessing - Embedding</span>

```python
round(x_train, 2)
```

```python
objective_col = ['22', '27', '30', '33']
numerical_col = ['1', '3', '4', '5', '6', '7', '10', '11', '13']

x_train = pd.get_dummies(x_train, columns=objective_col)
x_train.head()
```

```python
feature_col = []
for col in x_train.columns:
    feature_col.append(x_train[col])
```

```python
deepfm_base_model = keras.Sequential(
    keras.layers.DenseFeatures(feature_col.columns),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
)

deepfm_base_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

base_history = deepfm_base_model.fit(x_train, y_train, epochs=10, batch_size=32)
```

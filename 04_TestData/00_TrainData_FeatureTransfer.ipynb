```python
data.columns = [str(i) for i in range(1, len(data.columns)+1)]
data.columns
```

```python
top70_columns = ['6', '11', '7', '5', '13', '1', '3', '30', '10', '4', '22', '27', '8', '33', '15', '32', '17', '2', '38', '23', '25', '19', '9', '20', '37']
sorted_columns = sorted(top70_columns, key=int)
filtered_data = data[sorted_columns]
filtered_data
```

```python
filtered_data.info()
```

```python
filtered_data = ed.ValueLabelEncoder(filtered_data)
```

```python
imputed_data = ed.ImputeNan(filtered_data, strategy="median")
```

```python
skewed_data = ed.SkewColumns(imputed_data, first_range=0, last_range=12)
```

```python
rescaler_data = ed.NormalizedScaler(skewed_data, first_range=0, last_range=12, label=None)
```
